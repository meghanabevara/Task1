```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
df.to_csv("cleaned_titanic.csv", index=False)
```
